```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import pandas as pd
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\user\Desktop\AI_tasks\opencv-4.x\data\haarcascades\haarcascade_frontalface_default.xml")
```

```python
def show_image(image):
   plt.axis("off")
   plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
   plt.show()
```

```python
image = cv2.imread(r'C:\Users\user\Desktop\AI_tasks\image.jpg')
image.shape
```

```python
img=image[100:800, 0:1000]
gray =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image(img)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.2, 5)

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    face_gray = gray[y:y+h, x:x+w]
    face_color = img[y:y+h, x:x+w]
   
show_image(img)
```

### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks=pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv',sep=',')
drinks
```

### Step 4. Which continent drinks more beer on average?

```python
dg=drinks.groupby('continent')
group=dg.mean()
group
```

```python
group[group['beer_servings']==group['beer_servings'].max()].index
```

### Step 5. For each continent print the statistics for wine consumption.

```python
print("AFRICA statistics")
dg.get_group('AF')[['wine_servings']].describe()
```

```python
print("ASIA statistics")
dg.get_group('AS')[['wine_servings']].describe()
```

```python
print("EUROPE statistics")
dg.get_group('EU')[['wine_servings']].describe()
```

```python
print("OCEANIA statistics")
dg.get_group('OC')[['wine_servings']].describe()
```

```python
print("SOUTH AMERICA statistics")
dg.get_group('SA')[['wine_servings']].describe()
```
